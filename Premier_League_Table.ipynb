```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://www.premierleague.com/tables')
soup = BeautifulSoup(r.text,"lxml")
```

```python
teams = soup.find_all('tr', {'data-compseason': '210'})
football = []
for item in teams:
    nums = item.find_all('td')
    row={}
    row['Position'] = nums[1].find('span', {'class': 'value'}).text
    row['Club'] = nums[2].find('span', {'class': 'long'}).text
    row['Played'] = nums[3].text
    row['Won'] = nums[4].text
    row['Drawn'] = nums[5].text
    row['Lost'] = nums[6].text
    row['GF'] = nums[7].text
    row['GA'] = nums[8].text
    row['GD'] = nums[9].text
    row['Points'] = nums[10].text

    football.append(row)

frame = pd.DataFrame(football,columns=['Position','Club','Played','Won','Drawn','Lost','GF','GA','GD','Points'])
frame
```
